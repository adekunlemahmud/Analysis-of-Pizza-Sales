```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
orders = pd.read_csv(r"C:\Users\USER\Documents\Techcrush Python\orders.csv")
```

```python
pizza_types = pd.read_csv(r"C:\Users\USER\Documents\Techcrush Python\pizza_types.csv", encoding="latin1")

```

```python
pizza_types.head()
```

```python
pizzas = pd.read_csv(r"C:\Users\USER\Documents\Techcrush Python\pizzas.csv")
```

```python
pizzas.head()
```

```python
#joining the first two order tables

all_orders = pd.merge(order_details, orders, on="order_id", how="outer")

all_orders.head()
```

```python
#joining the last two pizza tables

all_pizza = pd.merge(pizza_types,pizzas, on="pizza_type_id", how="outer")

all_pizza.head()
```

```python
#joining all tables

df= pd.merge(all_orders,all_pizza, on="pizza_id", how="outer")
df.head()
```

```python
df.info()
```

```python
#change date to datetime data type
df["date"] = pd.to_datetime(df["date"])
```

```python
#calculate unit sales

df["sales"] = df["quantity"] * df["price"]
```

## No. 1 Total Revenues/Sales

```python

total_sales = df["sales"].sum().round(2)

total_sales
```

## No. 2 Quantity Sold

```python
total_quantity_sold= df["quantity"].sum()

total_quantity_sold
```

## No. 3 Total Orders

```python
total_orders = df["order_id"].nunique()

total_orders
```

## No. 4 Types of Pizzas sold

```python
types_of_pizza = df["name"].nunique()

types_of_pizza
```

## No. 5 Average price of pizzas

```python
average_pizza_price = df["price"].mean().round(2)

average_pizza_price
```

## No. 6 Peak hours of sales

```python
df["order_time"] = pd.to_datetime(df["time"]).dt.hour

peak_hours = df.groupby("order_time").size().sort_values(ascending = False)

peak_hours
```

```python
plt.bar(peak_hours.index, peak_hours.values)
plt.title("Graph of sales per hour")

plt.show()
```

## No. 7 Sales made in each day of the week 

```python
df["day_of_the_week"] = df["date"].dt.day_name()

sales_by_day = df.groupby("day_of_the_week")["sales"].sum()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sales_by_day = sales_by_day.reindex(day_order)

sales_by_day
```

## Day with the most sales

```python
peak_sales_day = sales_by_day.idxmax()
peak_sales = sales_by_day.max()

peak_sales_day, peak_sales
```

```python
plt.bar(sales_by_day.index, sales_by_day.values)
plt.xticks(rotation=45)
plt.title("Graph of sales per day")

plt.show()
```

## No. 8 5 Top selling pizzas

```python
top_selling_pizzas = df.groupby("name")["quantity"].sum().sort_values(ascending = False).head(5)

top_selling_pizzas
```

## No. 9 Sales made in each month

```python
df["year_month"] = df["date"].dt.to_period("M")

monthly_sales = df.groupby('year_month')['sales'].sum()

monthly_sales
```

## No. 10 Pizzas not doing well

```python
least_selling_pizzas = df.groupby("name")["quantity"].sum().sort_values().head(5)

least_selling_pizzas
```
